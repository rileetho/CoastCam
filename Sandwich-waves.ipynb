```python
# Open dataset
url ='https://dods.ndbc.noaa.gov/thredds/dodsC/data/stdmet/' + str(buoy) + '/' + str(buoy) + 'h9999.nc'
ds = xr.open_dataset(url)
ds
```

```python
lat1 = ds.latitude.values
lon1 = ds.longitude.values
print(lon1,lat1)
#ds = ds.sel(time=slice("2020-01-01","2023-06-01")) ####<<<<<We can us the datetime object to subset. no need to convert to strings. 

plt.figure()
ds.wave_height.plot()
plt.title('NDBC buoy observations')
plt.ylabel('Significant Wave height (m)')
plt.xlabel('YYYY-MM-DD, UTC')
fig_name = 'NDBC'+buoy + '_Hs.png'
#plt.savefig(fig_name)
```
